Count: The number of non-null or valid entries in the column. This tells you how many data points are available for that variable.
Mean: The arithmetic average of the values in the column. It is calculated by summing all the values and dividing by the count of values.
Std (Standard Deviation): A measure of how spread out the values are around the mean. A higher standard deviation indicates that the data points are more spread out from the mean, while a lower value means they are closer to the mean.
Min: The smallest value in the column.
25% (First Quartile or Q1): This is the value below which 25% of the data falls. It represents the lower bound of the middle 50% of the data.
50% (Median or Q2): The middle value when the data is sorted. Half of the data points fall below this value, and half fall above. This is often a better measure of central tendency when the data contains outliers.
75% (Third Quartile or Q3): The value below which 75% of the data falls. It represents the upper bound of the middle 50% of the data.
Max: The largest value in the column.
